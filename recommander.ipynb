```python
#Divide the data into training and test data
A_train,A_test=train_test_split(df,random_state=1)
print(len(A_train),len(A_test))
print(A_train.shape)
```

```python
#Build a recommendation model on training data
## Let use correlation between movies to define a similarity measure
##similarity matrix
A_train=A_train.fillna(0)
movie_similar=A_train.corr(method='pearson')
movie_similar.head(10)
```

```python
##Now let make recommandations #Make predictions on the test data

def get_similar_movie(movie_name,rating):
    similar_score=movie_similar[movie_name]*(rating)
    similar_score=similar_score.sort_values(ascending=False)
    
    return similar_score
```

```python
##Now let us test our model for the first user in the test set

user1=A_test.iloc[0,:]
```

```python
##let convert into an appropriate form
user1=[(mov[i],user1[i]) for i in range(206)]
```

```python

similar_movies=pd.DataFrame()
for movie,rating in user1:
    similar_movies=similar_movies.append(get_similar_movie(movie,rating),ignore_index=True)
    
print("the ranking for the first user in the test set are:")
similar_movies.sum().sort_values(ascending=False)
```

```python
##For the second
user2=A_test.iloc[1,:]
user2=[(mov[i],user2[i]) for i in range(206)]
similar_movies=pd.DataFrame()
for movie,rating in user2:
    similar_movies=similar_movies.append(get_similar_movie(movie,rating),ignore_index=True)
    
print("the ranking for the second user in the test set are:")
similar_movies.sum().sort_values(ascending=False)
```
